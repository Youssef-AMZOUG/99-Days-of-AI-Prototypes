```python
# --- STEP 1: INSTALL LIBRARIES ---
# We need 'accelerate' for modern models to run fast on GPU
!pip install transformers torch accelerate

from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# --- STEP 2: LOAD A SMARTER MODEL ---
# We use Qwen1.5-1.8B-Chat. It is MUCH smarter than DialoGPT.
# It can answer questions, code, and explain math.
model_name = "Qwen/Qwen1.5-1.8B-Chat"

print(f"Downloading {model_name}... (This handles math/code better)")
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"  # This automatically finds the GPU!
)

print("\n-----------------------------------")
print("SUCCESS: Model loaded on GPU!")
print("-----------------------------------\n")

# --- STEP 3: THE CHAT LOOP ---
messages = [] # We keep a list of messages so it remembers context

while True:
    user_input = input("\nYou: ")

    if user_input.lower() in ["quit", "exit"]:
        print("Chatbot: Goodbye!")
        break

    # Add the user's message to the history
    messages.append({"role": "user", "content": user_input})

    # Prepare the input for the model
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    # Move inputs to GPU
    model_inputs = tokenizer([text], return_tensors="pt").to(model.device)

    # Generate the response
    generated_ids = model.generate(
        model_inputs.input_ids,
        max_new_tokens=512,  # Allows for longer answers (algorithms/math)
        temperature=0.7      # 0.7 makes it creative but focused
    )

    # Extract only the new part of the response
    generated_ids = [
        output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
    ]

    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]

    # Add bot response to history so it remembers for next time
    messages.append({"role": "assistant", "content": response})

    print(f"Chatbot: {response}")
```
